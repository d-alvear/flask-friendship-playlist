```python
import pandas as pd

tracks = pd.read_csv('../data/interim/tracks_genre.csv', sep=";")
tracks.head()
```

```python
# dropping tracks without subgenres
idx = tracks[tracks['subgenres'] == "[]"].index

tracks.drop(idx, inplace=True)

tracks.reset_index(inplace=True, drop=True)

tracks.head()
```

```python
duplicates = tracks[tracks.duplicated(['track_name','artist']) == True].sort_values('track_name')
```

```python
grouped = tracks.groupby(['track_name','artist']).size().reset_index()
grouped.rename({0:'count'}, axis=1, inplace=True)
grouped.sort_values('count', ascending=False)
grouped = grouped[grouped['count'] > 1].sort_values('count', ascending=False)
grouped
```

```python
track_name = str(grouped.loc[28717,'track_name'])
artist = str(grouped.loc[28717,'artist'])

tracks[(tracks['track_name'] == track_name) & (tracks['artist'] == artist)]
```

```python
def handle_duplicates(grouped_df):
    keep = []
    for i,row in grouped_df.iterrows():

        duplicates = tracks[(tracks['track_name'] == str(row['track_name'])) & (tracks['artist'] == str(row['artist']))]
        max_popularity = duplicates['track_popularity'].max()
        
        for j,record in duplicates.iterrows():
            
            if ((record['track_popularity'] == max_popularity) and (isinstance(record['track_preview_url'], str))):
                keep.append(int(j))
                break
            
            elif ((isinstance(record['track_preview_url'], str)) and (record['track_popularity'] != max_popularity)):
                keep.append(int(j))
                break 

            else:
                idx = duplicates[duplicates['track_popularity'] == max_popularity].index
                keep.append(idx[0])
                break

    return keep
```

```python
keep_idx = handle_duplicates(grouped)
len(keep_idx)
```

```python
tracks_sorted = tracks.sort_values('track_preview_url')
tracks_sorted.drop_duplicates(['track_name','artist'], inplace=True)
tracks_sorted
```

```python
tracks_sorted.sort_index(inplace=True)
tracks_sorted.reset_index(drop=True, inplace=True)
tracks_sorted
```

```python
tracks_sorted.to_csv('../data/interim/tracks_no_duplicates.csv', sep=";", index=False)
```
